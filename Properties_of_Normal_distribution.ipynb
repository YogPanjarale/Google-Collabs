Sum of faces of 2 dice rolled 1000 times


```python
import random
import statistics as st
import plotly.figure_factory as ff
import plotly.graph_objects as go
```

```python
def rollDice():
  return random.randrange(1,7)
def sumOf2Rolls():
  return rollDice()+rollDice()
# doing this 1000 times
def listNtimes(n:int):
  l=[]
  for i in range(n):
    l.append(sumOf2Rolls())
  return l
l=listNtimes(10000)
print(l)
```

```python
mean = st.mean(l)
median = st.median(l)
mode = st.mode(l)
stdev = st.stdev(l)

print(f'Mean: {mean}\nMedian: {median}\nMode: {mode}\nStandard Deviation: {stdev}')
```

```python
fig = ff.create_distplot([l],["List"] ,show_hist=False)
fig.add_trace(go.Scatter(x=[mean,mean],y=[0,0.2],mode="lines",name="Mean"))
fig.show()
```

```python
stdStart,stdEnd=mean-stdev,mean+stdev
std2Start,std2End=mean- (stdev*2),mean+ (stdev*2)
std3Start,std3End=mean- (stdev*3),mean+ (stdev*3)
print(stdStart,stdEnd,std2Start,std2End,std3Start,std3End)
```
